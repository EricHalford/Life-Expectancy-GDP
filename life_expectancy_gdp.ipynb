```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
popData = pd.read_csv('all_data.csv')
print(popData.head())
print(popData['Country'].unique())
print(popData.columns)
```

```python
"""#Project Goals
- Identify if GDP is correlated to life expectancy and if so how strong is the relationship
- Analyze GDP growth / decline YoY and see if that has influenced life expectancy
"""
```

```python
# Seperating by country

Chile = popData.loc[popData['Country']=='Chile', label = 'Chile']
China = popData.loc[popData['Country']=='China']
Germany = popData.loc[popData['Country']=='Germany']
Mexico = popData.loc[popData['Country']=='Mexico']
USA = popData.loc[popData['Country']=='United States of America']
Zimbabwe = popData.loc[popData['Country']=='Zimbabwe']

```

```python
Countries = popData['Country'].unique()
'''for x in Countries:
    plt.plot(x['Year'], x['Life expectancy at birth (years)'])'''

plt.plot(Chile['Year'], Chile['Life expectancy at birth (years)'], label = 'Chile')
plt.plot(China['Year'], China['Life expectancy at birth (years)'], label = 'China')
plt.plot(Germany['Year'], Germany['Life expectancy at birth (years)'], label = 'Germany')
plt.plot(Mexico['Year'], Mexico['Life expectancy at birth (years)'], label = 'Mexico')
plt.plot(USA['Year'], USA['Life expectancy at birth (years)'], label = 'USA')
plt.title('Lifespan over the years')
#plt.plot(Zimbabwe['Year'], Zimbabwe['Life expectancy at birth (years)'])
#plt.legend(labels = Countries)
plt.legend()
plt.show()
plt.clf
```

```python
plt.plot(Chile['Year'], Chile['GDP'], label = 'Chile')
plt.plot(China['Year'], China['GDP'], label = 'China')
plt.plot(Germany['Year'], Germany['GDP'], label = 'Germany')
plt.plot(Mexico['Year'], Mexico['GDP'], label = 'Mexico')
plt.plot(USA['Year'], USA['GDP'], label = 'USA')
plt.legend()
plt.show()
plt.clf()
```

```python
print(Chile['GDP'].corr( Chile['Life expectancy at birth (years)'] ))
print(China['GDP'].corr( China['Life expectancy at birth (years)'] ))
print(Germany['GDP'].corr( Germany['Life expectancy at birth (years)']))
print(Mexico['GDP'].corr(Mexico['Life expectancy at birth (years)']))
print(USA['GDP'].corr( USA['Life expectancy at birth (years)']))
```
